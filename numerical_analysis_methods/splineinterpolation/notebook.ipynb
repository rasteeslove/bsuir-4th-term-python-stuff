## Interpolating $\frac{1}{x}$ on $[1; 2]$:

```python
a = 1
b = 2
nodes_num = 6

def hyperbole(x):
    return 1 / x
```

```python
xs = np.linspace(a, b, nodes_num)
ys = np.array([hyperbole(x) for x in xs])
nodes = [(x, y) for (x, y) in zip(xs, ys)]

print('Interpolation nodes:\n')
for (i, node) in enumerate(nodes):
    x = node[0]
    y = node[1]
    print(f'{i+1}) ({round(x, 4)}, {round(y, 4)})')
```

```python
spline_coeffs = methods.interpolate(nodes)

# a, b, c, d coefficients for each spline
print('Splines\' a, b, c and d coefficients:\n')
print(np.array(spline_coeffs))
```

```python
splines = methods.splines(nodes)

for spline in splines:
    methods.latex_print(spline)
```

```python
def plot_splines(xs, ys, splines, true_func=None):
    plt.scatter(xs, ys)
    nodes_num = len(nodes)

    spline_domains = [np.linspace(xs[i], xs[i+1], 20) for i in range(nodes_num - 1)]

    for (i, (spline, domain)) in enumerate(zip(splines, spline_domains)):
        plt.plot(domain, spline(domain - xs[i+1]))

    if true_func:
        full_domain = np.linspace(xs[0], xs[-1], 20*(nodes_num - 1))
        plt.plot(full_domain, true_func(full_domain), color='black', linestyle='dashed')

    plt.show()
```

```python
plot_splines(xs, ys, splines)
```

```python
plot_splines(xs, ys, [spline.deriv() for spline in splines])
```

```python
plot_splines(xs, ys, [spline.deriv(2) for spline in splines])
```

```python
plot_splines(xs, ys, splines, true_func=hyperbole)
```

***

## Evaluating interpolation in $x_0 = \frac{a+b}{2}$:

```python
x0 = (a+b) / 2

y01 = hyperbole(x0)
desired_spline = splines[2]
y02 = desired_spline(x0 - xs[3])

print(f'x0 = (a+b) / 2 = {x0}\n')
print(f'1/x at x={x0} is {y01}')
print(f'splines({x0}) = {y02}')
print(f'diff = {abs(y01-y02)}')
```

```python
desired_domain = np.linspace(1.49, 1.51, 100)
plt.plot(desired_domain, hyperbole(desired_domain), label='hyperbole')
plt.plot(desired_domain, desired_spline(desired_domain - xs[3]), label='spline')
plt.legend()
plt.show()
```
